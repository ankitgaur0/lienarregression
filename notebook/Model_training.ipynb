# now we have to make the pipelines for creating pipeline 
# before it done load data,drop the unnecessary columns, and make the variables for dependant and independant features.

```python
# for that i have to import the library of scikit-learn
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
# now we import the pipline and transformer module
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
numerical_pipeline=Pipeline(
    steps=[
        ("imputee",SimpleImputer()),
        ("scaling",StandardScaler())

    ]
)
categorical_pipeline=Pipeline(
    steps=[
        ("imputer",SimpleImputer(strategy="most_frequent")),
        ("encoder",OrdinalEncoder(categories=[cut_category,color_category,clarity_category]))
    ]
)
```

```python
columtrans=ColumnTransformer(
    [
        ("categorical",categorical_pipeline,categorical_columns),
        ("numberical",numerical_pipeline,numerical_columns)
    ]
)
```

```python
def check_accuracies(true_value,predict_value):
    mae=mean_absolute_error(true_value,predict_value)
    mse=mean_squared_error(true_value,predict_value)
    rmse=np.sqrt(mean_squared_error(true_value,predict_value))
    #r2score_value=r2_score(true_value,predict_value)*100

    return (mae,mse,rmse)
```
